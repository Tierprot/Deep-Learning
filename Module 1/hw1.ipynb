```python
import torch
from torch import utils
from torchvision import datasets, transforms
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
path='./MNIST_data'
```

```python
data = datasets.MNIST(path, train=True, download=True, transform=transforms.ToTensor())
print(data)
```

```python
data[0]
```

```python
def make_N_O(mnist_tensors_collection):
    #шаг 1 - агрегируем изображения 
    combine_tensors = lambda data: torch.stack([el[0] for el in data])
    all_tensors_combined = combine_tensors(mnist_tensors_collection)
    print(all_tensors_combined.shape)
    
    #шаг 2 - получаем параметры для нормализации
    mean, std = all_tensors_combined.mean(), all_tensors_combined.std()
    print(f'dataset mean: {mean}, std: {std}')
    
    #шаг 3 - формируем сходную с изначальной коллекцию кортежей вида (тензор, соответствующее число)
    transformed_collection = [(transforms.Normalize((mean,), (std,))(tensor), label) for 
                                      tensor, label in mnist_tensors_collection]
    
    #step 4 - проверяем, что распределение в полученной коллекции более-менее соответсвтует ожидаемому
    transformed_collection_tensors = combine_tensors(transformed_collection)
    print(f'modified dataset mean: {transformed_collection_tensors.mean()}, std:{transformed_collection_tensors.std()}')
    return transformed_collection
```

```python
new_data = make_N_O(data)
```
