```python
import torch
from torch import utils
from torchvision import datasets, transforms
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
path='./MNIST_data'
```

```python
def custom_transforms(dataset):
    tensors = dataset.data.float().div(255)
    return transforms.Compose([
                transforms.ToTensor(),
                transforms.Normalize((tensors.mean(),), (tensors.std(),)),
           ])
```

```python
data = datasets.MNIST(path, train=True, download=True)
```

```python
data.transform = custom_transforms(data)
```
